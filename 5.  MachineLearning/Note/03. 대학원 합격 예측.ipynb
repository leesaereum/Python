```python
import pandas as pd
data = pd.read_csv("../Data/gpascore.csv")
data.head()
```

admit : 합격여부    
gre : 영어성적    
gpa : 학점    
rank : 지원 대학원 수준

# Preprocessing(전처리)

```python
# 결측치 처리
data.isna().sum()
```

```python
# 결측치 데이터 삭제
data.dropna(inplace=True)

# 결측치 데이터를 다른 값으로 대체
data.fillna(100)
```

```python
# 결측치 삭제 확인
data.isna().sum()
```

```python
# rank의 종류 파악
sorted(data['rank'].unique())
```

```python
# 영어 점수의 최소값과 최대값
print('최대값 :',data['gre'].max())
print('최대값 :',data['gre'].min())
```

```python
# Training Data
x = data[['gre','gpa','rank']]
x.head()
```

```python
# target Data
y = data['admit']
y.head()
```

```python
# 데이터 갯수 확인
x.count()
```

---
# Deep Learning Model 만들기

```python
from tensorflow import keras
```

```python
model = keras.Sequential()
# 1번째 Hidden Layer : 256 노드 (2의 배수, 임의의 갯수(Hyper Parameter))
# activation = sigmoid, tanh, relu, softmax, leakyRelu
model.add(keras.layers.Dense(256, activation='tanh', input_shape=(3,)))
model.add(keras.layers.Dense(1, activation='sigmoid'))
```

```python
# 손실함수
model.compile(optimizer='adam',loss='binary_crossentropy', metrics='accuracy')
# Opimaizer : learning rate를 알맞게 조정해줌
# - adam
# - adgred
# - adadelta
# - rmsprop
# - sgd
# loss : 손실함수
# - binary-crossentropy : 분류 및 확률 예측 
# metrics : 모델의 평가 기준
```

```python
# 학습시키기
import numpy as np
model.fit(np.array(x), np.array(y), epochs=500)
```

```python
# 예측하기

model.predict(
    [
        [700, 3.7, 1],
        [700, 3.7, 2],
        [700, 3.7, 3],
        [700, 3.7, 4],
    ]
)
```
