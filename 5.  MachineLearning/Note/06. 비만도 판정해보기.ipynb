### Data 정규화

```python
# height, weight 
csv['height'] = csv['height'] / csv['height'].max()
csv['weight'] = csv['weight'] / csv['weight'].max()
```

```python
csv.head()
```

```python
# label 
bclass = {'thin':[1,0,0], 'normal':[0,1,0], 'fat':[0,0,1]}
y = []
for i,v in enumerate(csv['label']):
    y.append(bclass[v])
y
```

```python
# Training 과 Test로 나누기
X_train = csv[['height','weight']][0:15000]
y_train = y[0:15000]
X_test = csv[['height','weight']][15000:]
y_test = y[15000:]
```

```python
# 길이 확인
print(len(X_train))
print(len(y_train))
print(len(X_test))
print(len(y_test))
```

---
# Deep Learning

```python
from tensorflow import keras
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(512, activation='relu',input_shape=(2,)))
model.add(keras.layers.Dense(3, activation='softmax'))
```

```python
model.compile(
    optimizer = 'sgd',
    loss='categorical_crossentropy',
    metrics = ['accuracy']
)
```

```python
# 데이터 훈련
import numpy as np
model.fit(
    np.array(X_train),
    np.array(y_train),
    epochs=500
)
```

```python
# 테스트 데이터로 평가하기
score = model.evaluate(X_test,np.array(y_test))
print('loss :',score[0])
print('accuracy :',score[1])
```

> optimizer를 고르는 기준은은 평가시 loss가 적은 순
