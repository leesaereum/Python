admit : 합격여부    
gre : 영어성적

rank : 지원 대학원 수준

# Preprocessing(전처리)

```python
# 결측치 처리
data.isna().sum()
```

```python
# 결측치 데이터 삭제
data.dropna(inplace=True)

# 결측치 데이터를 다른 값으로 대체
data.fillna(100)
```

```python
# 결측치 삭제 확인
data.isna().sum()
```

```python
# rank의 종류 파악
sorted(data['rank'].unique())
```

```python
# 영어 점수의 최소값과 최대값
print('최대값 :',data['gre'].max())
print('최대값 :',data['gre'].min())
```

```python
# Training Data
x = data[['gre','gpa','rank']]
x.head()
```

```python
# target Data
y = data['admit']
y.head()
```

```python
# 데이터 갯수 확인
x.count()
```

---
# Deep Learning Model 만들기

```python
from tensorflow import keras
```

```python
input_layer = keras.layers.InputLayer(input_shape=(3,))
hidden_layer = keras.layers.Dense(256,activation='tanh')
output_layer = keras.layers.Dense(1,activation='sigmoid')
```

```python
model = keras.Sequential(
    [
        input_layer,
        hidden_layer,
        output_layer
    ]
)
```

```python
# 손실함수
model.compile(optimizer='adam',loss='binary_crossentropy', metrics='accuracy')
# Opimaizer : learning rate를 알맞게 조정해줌
# - adam
# - adgred
# - adadelta
# - rmsprop
# - sgd
# loss : 손실함수
# - binary-crossentropy : 분류 및 확률 예측 
# metrics : 모델의 평가 기준
```

```python
# 학습시키기
import numpy as np
model.fit(np.array(x), np.array(y), epochs=500)
```

```python
# hidden layer의 가중치 값 출력
hidden_layer.get_weights()[0]
```

```python
hidden_layer.get_weights()[1]
```

```python
# hidden layer의 출력 확인하기
intermediate_layer_model = keras.Model(inputs=model.input,outputs=model.layers[0].output)
intermediate_output = intermediate_layer_model(np.array(x))
```
