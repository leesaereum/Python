```python
import folium
import pandas as pd
```

```python
kyochon = pd.read_csv("../Data/Kyochon_map.csv")
pizzaschool = pd.read_csv("../Data/pizzaschool_map.csv")
```

```python
map = folium.Map(
    location=[37.5502 , 126.982],
    zoom_start=10.5,
)

for i in range(len(kyochon)):
    txt = '<h4>' +'교촌치킨' + '</h4>'
    txt += '<p>'+ kyochon['가맹점명'].iloc[i] + '</p>'

    popup = folium.Popup( txt ,min_width=100, max_width=100 )
    folium.CircleMarker(
            [kyochon.iloc[i]['위도'], kyochon.iloc[i]['경도']],
            radius=10,
            color='orange',
            fill_color='orange',
            fill_opacity= 0.4,
            popup = popup,
            ).add_to(map)
            
for i in range(len(pizzaschool)):
    txt = '<h4>' +'피자스쿨' + '</h4>'
    txt += '<p>'+ pizzaschool['가맹점명'].iloc[i] + '</p>'

    popup = folium.Popup( txt ,min_width=100, max_width=100 )
    folium.CircleMarker(
            [pizzaschool.iloc[i]['위도'], pizzaschool.iloc[i]['경도']],
            radius=10,
            color='pink',
            fill_color='pink',
            fill_opacity= 0.4,
            popup = popup,
            ).add_to(map)

# map.save('../Data/Kochon_Pizzaschool_map.html')
map
```
