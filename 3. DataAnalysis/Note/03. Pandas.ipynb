```python
df
```

```python
# index가 서준인 행을 선택
print(df.loc['서준'])
print("")
print(df.iloc[0])
```

### 행과 열을 이용한 데이터 선택

```python
# index이름을 정하기
aa = {'수학':[90,80,70], '영어':[98,89,95], '음악':[85,95,100], '체육':[100,90,90],'이름':['서준','우현','인아']}
df = pd.DataFrame(aa)
df
```

```python
df.set_index('이름',inplace=True)
df
```

```python
print(df.loc['서준',['음악','체육']])
print(df.loc['서준','음악':'체육'])
print(df.loc['서준','음악':])
print(df.iloc[0,[2,3]])
print(df.iloc[0,2:4])
print(df.iloc[0,2:])
```

```python
df['국어'] = [10,20,30]
df
```

```python
df.loc['철수'] = [10,20,30,40,50]
df
```

```python
df.loc['영희'] = df.loc['인아']
df
```

```python
# 서준의 체육점수를 80점으로 변경하기
df.loc['서준','체육'] = 80
df.iloc[0,3] = 80
df
```

```python
# 서준의 음악 점수를 100점, 체육 점수를 50점으로 변경하기
df.loc['서준','음악':'체육'] = 100,50
df.iloc[0,2:4] = [100,50]
df
```

```python
# 행과 열의 위치바꾸기 
df.transpose() # = df.T
```

```python
aa = {'c0':[1,2,3],'c1':[4,5,6],'c2':[7,8,9],'c3':[10,11,12],'c4':[13,14,15]}
df = pd.DataFrame(aa,index=['r0','r1','r2'])
df
```

```python
# 인덱스를 재설정하기
new_index = ['r0','r1','r2','r3','r4']
ndf = df.reindex(new_index,fill_value=0)
ndf
```

```python
# 내림차순으로 행 인덱스 정렬
df.sort_index(ascending=False)
```

```python
# 특정 값을 기준으로 정렬
df.sort_values(by='c1',ascending=False,inplace=True)
df
```

# DataFrame 연산

```python
import seaborn as sns
```

```python
titnanic = sns.load_dataset('titanic')
titnanic.head()
```

```python
# titanic에서 age와 fare부분의 컬럼만 발췌하여 df라는 dataFrame 만들기
df = titnanic.loc[:,['age','fare']]
df.head()
```

```python
# 모든 컬럼을 10씩 증가하여 addtion이라는 DataFrame 생성하기
addition = df + 10
addition.head()
```

```python
# addition과 df의 차를 구하여 subtraction이라는 Dataframe 만들기
subtraction = addition - df
subtraction
```
