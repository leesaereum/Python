```python
# train.csv를 읽어서 train이라는 변수에 넣기 (단 index는 PassengerId를 사용한다)
import pandas as pd
train = pd.read_csv("../Data/train.csv",index_col='PassengerId')
train.head()
```

```python
# 행과 열의 갯수를 출력하기
train.shape
```

```python
# train의 Survived 컬럼의 앞에서부터 5줄 출력하기
print(train.Survived[:5])
```

```python
# train의 Sex, Pclass, Survived 컬럼의 앞에서부터 5줄 출력하기
print(train.loc[:5,['Sex','Pclass','Survived']])
```

```python
# 1번부터 3번 승객까지의 데이터를 가져오기
train.loc[:3,]

# 1,3,7,13번의 승객 데이터를 가져오기
train.loc[[1,3,7,13],:]

# 1,3,7,13번의 승객들의 Pclass, Sex, Survived 가져오기
train.loc[[1,3,7,13],['Sex','Pclass','Survived']]
```

```python
# 성별이 남자인 데이터만 가져오기
train[train['Sex']=='male']
```

```python
# 운임요금이 20달러 이상인 데이터중 처음부터 10개만 출력하기
train[train['Fare']>=20].head(10)
```

```python
# 탑승항구가 S항구 Q항구인(* C항구가 아닌)데이터만 출력하기
train[train['Embarked'] != 'C']
```
