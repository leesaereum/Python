# 데이터 사전처리(Preprocessing)
- 머신러닝 등 데이터 분석의 정확도는 분석 데이터의 품질에 좌우된다.
- 데이터의 품질을 높이기 위해서는 누락 데이터, 중복 데이터 등 오류를 수정하고 분석목적에 맞게 변형하는 과정이 필요
- 수집한 데이터를 분석 목적에 적합하도록 하는 작업을 사전 처리(Prepreocessing)이라고 한다.

### 누락 데이터(NaN: Not a Number)
- 데이터를 파일로 입력할때 빠트리거나 파일 형식을 변환하는 과정에서 데이터가 손실되는 경우가 발생
- 머신러닝 분석 모형에 데이터를 입력하기 전에 반드시 누락 데이터를 제거하고 다른 적절한 값으로 대체 하여야 한다.
- 누락 데이터가 많아지면 데이터의 품질이 떨어지고 머신러닝 분석 알고리즘을 왜곡하는 현상이 발생한다.

# 범주형(카테고리) 데이처 처리
- 연속 데이터를 그대로 사용하기 보다는 일정한 구간(binning)으로 나눠서 분석하는것이 효율적인 경우가 많다

```python
# horsepower 3개의 범위로 나누기 
count, bin_dividers = np.histogram(df['horsepower'], bins=3)
bin_dividers
```

```python
# 3개의 bin에 이름 지정
bin_names = ['저출력','보통출력','고출력']
df['hp_bin'] = pd.cut(
    x=df['horsepower'],
    bins=bin_dividers,
    labels = bin_names,
    include_lowest = True #첫 경계값 포함
)
```

```python
df[['horsepower','hp_bin']].head(25)
```

### One-Hot Encoding
- 범주형 데이터를 0,1을 갖는 원핫벡터로 변환
- 결과는 선형대수학에서 정의하는 희소행렬(sparse matrix)로 변환한다.
- 희소행렬은 (행,열)좌표와 값 형태로 정리

```python
from sklearn import preprocessing
```

```python
# 전처리를 위한 encoder 객체 만들기
label_encoder = preprocessing.LabelEncoder() # label encoder 생성
onehot_encoder = preprocessing.OneHotEncoder() # one hot encoder 생성  
```

```python
# label encoder로 문자열 범주를 숫자형 범주로 변환
onehot_labeled = label_encoder.fit_transform(df['hp_bin'].head(15))
onehot_labeled
```

```python
# 2차원 행렬로 형태를 변경
onehot_reshape = onehot_labeled.reshape(len(onehot_labeled),1)
onehot_reshape
```

```python
# 희소행렬 
onehot_fitted = onehot_encoder.fit_transform(onehot_reshape)
print(onehot_fitted)
```

### 정규화(Normalization)
- 각 열에 있는 숫자 데이터의 상대적 크기 때문에 머신러닝 분석결과가 달라 질 수 있다.
- 숫자 데이터의 상대적인 크기 차이를 제거

```python
# 최대값의 절대값을 이용한 정규화
df.horsepower.describe()
```

```python
# horsepower 열의 최대값의 절대값으로 모든 데이터를 나누기 하여 저장
horsepower1 = df.horsepower / abs(df.horsepower.max())
horsepower1.describe()
```

```python
# 각 열의 데이터중에 최대값과 최소값을 뺄셈한 값으로 나누기 하여 저장
df.horsepower.describe()
```

```python
# 각 열에서 해당열의 최소값을 뺀 값을 분자로 하고 
min_x = df.horsepower - df.horsepower.min()

# 해당 열의 최대값과 최소값의 차를 분모로 하는 수를 계산하면 가장 큰 값은 1이 된다.
min_max = df.horsepower.max() - df.horsepower.min()

horsepower2 = min_x/min_max
horsepower2.describe()
```

### Timestamp를 Period로 변환

```python
# 날짜 형식의 문자열로 구성되는 리스트 정의
dates = ['2019-01-01','2020-03-01','2021-06-01']
```

```python
ts_dates = pd.to_datetime(dates)
ts_dates
```

```python
pr_day = ts_dates.to_period(freq = 'D')
print(pr_day)
pr_month = ts_dates.to_period(freq = 'M')
print(pr_month)
pr_year = ts_dates.to_period(freq = 'Y')
print(pr_year)
```

> freq Option   
> D (Day: 1일)   
> W (Week:1주)   
> M (Month end: 월말)   
> MS (Month begin: 월초)   
> Q (Quarter end : 분기말)   
> QS (Quarter begin: 분기초)   
> A (year end : 연말)   
> AS (year begin: 연초)   
> B (Business day : 휴일 제외)   
> H (hour : 1시간)   
> T (minute :1분)   
> S (second: 1초)

# 시계열 데이터 만들기

```python
# Timestamp의 배열 만들기 : 월 간격, 월의 시작일 기준
ts_ms = pd.date_range(
    start='2022-01-01', # 날짜 범위의 시작
    end = None, # 날짜 범위의 끝
    periods = 6, # 생성할 Timestampe의 갯수
    freq='MS', # 시간 간격
    tz = 'Asia/Seoul' # 시간대
)
ts_ms
```

```python
# Timestamp의 배열 만들기 : 월 간격, 월의 마지막날 기준
ts_me = pd.date_range(
    start='2022-01-01', # 날짜 범위의 시작
    end = None, # 날짜 범위의 끝
    periods = 6, # 생성할 Timestampe의 갯수
    freq='M', # 시간 간격
    tz = 'Asia/Seoul' # 시간대
)
ts_me
```

```python
# Timestamp의 배열 만들기 : 분기 간격, 월의 마지막날 기준
ts_q = pd.date_range(
    start='2022-01-01', # 날짜 범위의 시작
    end = None, # 날짜 범위의 끝
    periods = 6, # 생성할 Timestampe의 갯수
    freq='Q', # 시간 간격
    tz = 'Asia/Seoul' # 시간대
)
ts_q
```
